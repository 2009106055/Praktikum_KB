<h1> MASUKKAN INFORMASI DIATAS KEDALAM DICTIONARY DENGAN KEY SESUAI NAMA INFORMASI TERSEBUT </h1>

```python
totalRecord = len(df)
namaAttribute = df.columns
jumlahObjek = len(df.select_dtypes(include=['object']).columns)
jumlahAngka = len(df.select_dtypes(include=['int64', 'float64']).columns)
jumlahNULL = len(df.columns[df.isna().sum() > 0])

mydictionary = {
    "Jumlah Total Record": totalRecord,     
    "Nama Attribute": namaAttribute,        
    "Jumlah Objek": jumlahObjek,            
    "Jumlah Angka": jumlahAngka,            
    "Jumlah NULL": jumlahNULL,         
}

pd.Series(mydictionary)
```

<h1> RUBAH NILAI SALAH SATU ATTRIBUTE BERTIPE ANGKA MENJADI DARI DESIMAL MENJADI BILANGAN BULAT </h1>

Dikarnekan ada error [cannot convert non-finite values (na or inf) to integer] Yang mungkin disebab kan adanya nilai Null pada atribute, maka saya buat dua cara

1. [errors='ignore'] yang dimana error akan di abaikan dan akan mejalan kan perintah tampa menegeluarkan peringatan error

2. Metode yang saya temukan di Stack Overflow yang mengguankan 'fillna function' yang dimana akan mengubah value Null dengan value yang sudah ditentukan

<h1> METODE [errors='ignore'] </h1>

```python
df["runtime"] = df["runtime"].astype('int64', errors='ignore')
df["runtime"].dtypes
```

```python
df["runtime"] = df["runtime"].astype('float64', errors='ignore')
```

<h1> METODE [fillna function] </h1>

```python
df = df.fillna(0)
df.runtime.astype('int64')
```
