<h5>Import Library/Package</h5>

```python
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
```

<h5>Import Database Yang Kita Punya</h5>

```python
dataset = pd.read_csv('FIFA22_PlayerCards.csv')
dataset
```

<h5>preprocessing data Di sini saya mencoba menggunakan Encoding untuk mengganti nilai-nilai bertipe data objek menjadi tipe data angka</h5>

```python
varEncoder = OrdinalEncoder()
dataset['Foot'] = varEncoder.fit_transform(dataset[['Foot']])
dataset['AWR'] = varEncoder.fit_transform(dataset[['AWR']])
dataset['DWR'] = varEncoder.fit_transform(dataset[['DWR']])
dataset['Position'] = varEncoder.fit_transform(dataset[['Position']])
```

```python
dataset
```

<h5>Menentukan record dan label</h5>

```python
x = dataset[['Foot','Position','AWR','DWR','OVR']]
y = dataset[['SM']]
```

<h5>Membagi data train pada variabel independen menjadi data train dan data test. Di sini saya menggunakan proporsi 70:30 (70% Data train dan 30% Data test)</h5>

```python

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0, test_size=.3)
```

<h5>menampilkan hasil x_train,x_test,y_train,y_test</h5>

```python
print(f'panjang x train : {len(x_train)}',
f'panjang y train : {len(y_train)}',
f'panjang x test : {len(x_test)}',
f'panjang y test : {len(y_test)}')
```

<h5>Di sini saya menetapkan jumlah kelas sebesar 2 (sesuaikan dengan jumlah kelas yang dimiliki dataset)<h5>

```python
num_class = 6
y_train = tf.keras.utils.to_categorical(y_train, num_class)
y_test = tf.keras.utils.to_categorical(y_test, num_class)
```

<h5>Buat model sequential dengan total layer sebanyak 5 layer menggunakan aktivasi relu pada 4 layer pertama dan aktivasi sigmoid pada layer terakhir, lalu tampilkan model tersebut dengan fungsi summary()</h5>

```python
seq = tf.keras.Sequential([
    tf.keras.layers.Dense(16, activation='relu', input_dim=5),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(num_class, activation='sigmoid')
])
seq.summary()
```

<h5>Lakukan Proses Compile</h5>

```python
seq.compile(
    optimizer=tf.keras.optimizers.Adam(),
    loss=tf.keras.losses.categorical_crossentropy,
    metrics=['accuracy']
)
```

<h5>Lakukan proses model training dengan fit</h5>

```python
accuracy = seq.fit(
    x = x_train,  y = y_train,
    validation_data = (x_test, y_test),
    epochs = 50,
    steps_per_epoch = 250,
    verbose = 2 ## tampilan compile per epoch 1 default, 2 tidak melihatkan perjalanan 0 langsung 
)
```

<h5>Melakukan Model Save dan menyimpan nya ke dalam format hdf5</h5>

```python
seq.save('model/seq_1.h5')
```
