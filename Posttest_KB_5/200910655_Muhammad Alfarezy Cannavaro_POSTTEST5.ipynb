<h1>DATA CLEANING</h1>

```python
dataset_cl1 = df.copy()
dataset_cl2 = df2.copy()
dataset_cl3 = df.copy()
```

```python
imputer_mean = SimpleImputer(strategy='mean')
imputer_median = SimpleImputer(strategy='median')
imputer_modus = SimpleImputer(strategy='most_frequent')
```

<h5>Modus</h5>

```python
dataset_cl1['runtime'] = imputer_modus.fit_transform(dataset_cl1[['runtime']])

dataset_cl1.isna().sum()
```

<h5>Mean</h5>

```python
dataset_cl2['Winner'] = imputer_mean.fit_transform(dataset_cl2[['Winner']])

dataset_cl2.isna().sum()
```

<h1>KNN (K-NEAREST NEIGHBOUR) CLASSIFICATION</h1>

```python
dataset_cl3.head()
```

```python
dataset_cl3.info()
```

```python
x = dataset_cl3[['runtime','vote_average','vote_count','popularity']]
y = dataset_cl3['status']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0, test_size=.2)
```

```python
classifier = KNeighborsClassifier(n_neighbors = 5)
```

```python
classifier.fit(x_train,y_train)
```

```python
predicted_value = classifier.predict(x_test)
accuracy = accuracy_score(y_test, predicted_value)
```

```python
print(f'hasil prediksi: {predicted_value}',
     f'Class: {y_test}',
     f'akurasi: {accuracy}',
     sep = '\n'
)
```

```python
classifier.score(x_test,y_test)
```

<h1>NAIVE BAYES CLASSIFICATION</h1>

```python
x = dataset_cl3[['runtime','vote_average','vote_count','popularity']]
y = dataset_cl3['status']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=1, shuffle=False, stratify=None)
```

```python
classifier = GaussianNB()
```

```python
classifier.fit(x_train,y_train)
```

```python
predicted_value = classifier.predict(x_test)
accuracy = accuracy_score(y_test, predicted_value)
```

```python
print(f'hasil prediksi: {predicted_value}',
     f'Class: {y_test}',
     f'akurasi: {accuracy}',
     sep = '\n'
)
```

```python
classifier.score(x_test,y_test)
```
