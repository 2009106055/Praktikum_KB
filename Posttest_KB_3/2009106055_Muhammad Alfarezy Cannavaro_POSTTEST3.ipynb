<h1> MENGIMPORT LIBRARY PANDAS,MATPLOTLIB,DAN SEABORN </h1>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

<h1>MEMBACA DATASET</h1>

```python
df = pd.read_csv("FIFA22_PlayerCards.csv")
```

```python
df
```

<h1>MELIHAT BANYAK DARI ISI RECORD</h1>

```python
df['AWR'].value_counts()
```

<h1>GRAFIK BATANG</h1>

Grafik yang di gunakan untuk perbandingan kategorikal,disini saya melihat visualisasi data yang dimana atribute yang di gunakan adalah atribute dari penggunaan kaki dari player tersebut dan kekuatan dari shooting player tersebut

```python
x = df['Foot']
y = df['SHO']
```

```python
plt.bar(x, y,
       align="edge", width=0.5
       )
plt.title("Grafik Batang")
plt.xlabel("Kaki Dari Player")
plt.ylabel("Kekuatan Tendangan")
plt.show()
```

<h1>GRAFIK PIE</h1>

Grafik Pie cocok digunakan pada data categorical. Grafik ini biasa digunakan untuk melihat distribusi, disini saya melihat atribute attack awarnes para pemain yang ada di dalam dataset saya

```python
df = df.dropna(subset=['AWR'])
```

```python
data_pie = df['AWR'].value_counts().rename_axis('AWR').reset_index(name='Jumlah')
```

```python
data_pie.head(10)
```

```python
plt.figure(figsize=(8,8), dpi=300)
plt.title("GRAFIK PIE AWARNESS")
plt.pie(df['AWR'].value_counts(), labels=data_pie['AWR'], autopct='%.1f%%', startangle=90)
plt.show()
```

<h1>GRAFIK SCATTER PLOT</h1>

diagram grafis yang dibangun dari dua sumbu X dan Y (variabel X dan variabel Y). Scatter plot menunjukan hubungan antara dua variabel,disini saya melihat overal dari masing2 player yang ada di dalam dataset saya

```python
sns.scatterplot(data = df, x='Position', y = 'OVR')
plt.title("PESEBARAN DATA ANTARA POSISI DARI PLAYER DAN OVERALL PLAYER TERSEBUT")
plt.show()
```
